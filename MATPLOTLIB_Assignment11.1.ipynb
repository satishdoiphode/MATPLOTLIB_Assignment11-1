```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
titanic.groupby('sex').count()
```

```python
a=titanic.groupby('sex').size()
plt.axis('equal')
plt.tight_layout()
a.plot(kind='pie',autopct='%.2f',fontsize=14,title='Male / Female proportion',legend=False,labels=["Males","Females"])
plt.ylabel("")
plt.show()
```

```python
x = titanic['age']
y=titanic['fare']
colors=np.where(titanic.sex=='female','red','blue')
plt.scatter(x,y,c=colors,alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
